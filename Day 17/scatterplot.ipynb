```python
# Convert 1D array with 8 elements to 3D array with 2x2 elements
import numpy as np

# 1D array with 8 elements
array_1d = np.arange(8)

# Reshape the 1D array to 3D array with shape (2, 2, 2)
array_3d = array_1d.reshape(2, 2, 2)
array_3d
```

```python
# Convert the array into a 1D array
# Flatten the 3D array back to 1D array
array_1d_again = array_3d.flatten()
array_1d_again
```

```python
# Use the scatter() method to draw a scatter plot diagram
import matplotlib.pyplot as plt

# Sample data for scatter plot
x = np.random.rand(10)
y = np.random.rand(10)

# Scatter plot
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
# A scatter plot with 1000 dots
# Generating 1000 random points
x_large = np.random.rand(1000)
y_large = np.random.rand(1000)

# Scatter plot with 1000 dots
plt.scatter(x_large, y_large)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with 1000 Dots')
plt.show()
```
