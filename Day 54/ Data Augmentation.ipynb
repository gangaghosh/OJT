### 2. Data Augmentation 
Implement data augmentation on a given image dataset using Keras. Show at least three 
different augmentation techniques and explain how they help improve model 
performance.

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python

# Load an example image
img = load_img('img.jpg') 
x = img_to_array(img)
x = np.expand_dims(x, axis=0)
```

```python

# Create an ImageDataGenerator and specify augmentation techniques
datagen = ImageDataGenerator(
    rotation_range=40,  
    width_shift_range=0.2, 
    height_shift_range=0.2,  
    horizontal_flip=True,  
    fill_mode='nearest'  
)
```

```python

# Generate augmented images
augmented_images = datagen.flow(x, batch_size=1)

# Display original and augmented images
plt.figure(figsize=(12, 6))
```

```python

# Original image
plt.subplot(1, 4, 1)
plt.imshow(img)
plt.title("Original Image")

```

```python

# Augmented images
for i in range(3):
    plt.subplot(1, 4, i + 2)
    batch = augmented_images.next()
    img_augmented = array_to_img(batch[0])
    plt.imshow(img_augmented)
    plt.title(f"Augmented Image {i+1}")

plt.show()
```
